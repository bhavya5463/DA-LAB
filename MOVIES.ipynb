```python
import pandas as pd
from dateutil.parser import parse
df=pd.read_csv('/content/movies_1_.csv')
df
```

```python
#pre-processing
import pandas as pd
from dateutil.parser import parse
df1=df.copy()
for i in range(len(df['Release Date'])):
  if(df['Release Date'][i]=='TBD'):
    df1=df1.drop(i)
  else:
    df1['Release Date'][i]=parse(df['Release Date'][i]).date()
df1.index=range(len(df1))
df1['Release Date']=pd.to_datetime(df1['Release Date'])
df1['year']=df1['Release Date'].dt.year
df1['month']=df1['Release Date'].dt.month
df1['day']=df1['Release Date'].dt.day
df1
```

```python
#1.	 (i). Find out the no of movies released in every month of the year 1995.
g=df1[df1['year']==1995]
k=g.groupby('month')['Title'].count()
k
```

```python
#(ii). Find out the no. of movies released in every year from 1990 to 1998
k=df1[(df1['year']>=1990) & (df1['year']<=1998)]
g=k.groupby('year')['Title'].count()
g
```

```python
#2.	(i). Find no. of movies released under each genre given in the database.
k=df1.groupby('Major Genre')['Title'].count()
k
```

```python
#(ii). Find the movies under each genre with 1MDB rating >7 and rotten tomatoes rating > 60.
k=df1[(df1['IMDB Rating']>7) & (df1['Rotten Tomatoes Rating']>60)]
g=k.groupby(['Major Genre','Title'])['Title'].count()
g
```

```python
#3.	(i). Find the movies released under each fiction with each director in the ascending order of release dates.
k=df1.sort_values(by='Release Date',ascending=True)
g=k.groupby(['Creative Type','Director','Title'])['Title'].count()
g
```

```python
#(ii). Find movies released under each distributor in the order of genre and director
k=df1.sort_values(by=['Major Genre','Director'],ascending=True)
g=k.groupby(['Distributor','Title'])['Title'].count()
g
```

```python
#4.	(i). Find the movies released world-wide and find out the revenue received world-wide other than US with their ratings.
df1['Worldwide Gross']=pd.to_numeric(df1['Worldwide Gross'],errors='coerce')
df1['US Gross']=pd.to_numeric(df1['US Gross'],errors='coerce')
df1['revenue']=df1['Worldwide Gross']-df1['US Gross']
df1[['Title','IMDB Rating','revenue']]
```

```python
#(ii). Find the movies with loss & profit released in each year with genre and ratings.
df1['Production Budget']=pd.to_numeric(df1['Production Budget'],errors='coerce')
df1['p and l']=df1['Production Budget']-df1['Worldwide Gross']
p=df1.groupby('year')
for i,j in p:
  print("year : ",i)
  print(j[["Title","Major Genre","IMDB Rating","p and l"]])
  print("\n")
```
