```python
import pandas as pd
df=pd.read_csv("/content/National_Universities_Rankings.csv")
df
```

```python
#1)Find the universities in which undergraduate students were admitted.
df["Name"][df["Undergrad Enrollment"].isnull()==False]
```

```python
#2)List the states along with the cities in which universities location
t=pd.DataFrame(df["Location"])
t
```

```python
#3.	List the cities &universities under each state.
l=[]
loc=[]
p=list(df["Location"])
for i in p:
  a=i.split(",")[1]
  b=i.split(",")[0]
  l.append(a)
  loc.append(b)
df['state']=l
df['city']=loc
df[['city','state','Name']]
```

```python
#4.	How many universities in each state have ranking<50? Print them along with ranking
a=df[df['Rank']<50].groupby(['state','Name'])['Rank'].apply(list)
p=a.reset_index(name='Rank')
p
```

```python
#5.	How many universities have both out-of state and in-state students?
df[df['In-state'].notnull() & df['Tuition and fees'].notnull()]['Name'].count()
```

```python
#6.	How many universities have marginal difference <=$5000 in in-state &out-of state tuition fees.
k=df['Tuition and fees'].str.replace(",","").str.split("$",expand=True).fillna(0)
m=k[1].astype('int64')
l=df['In-state'].str.replace(",","").str.split("$",expand=True).fillna(0)
n=l[1].astype('int64')
df[abs(m-n)<=5000]
```

```python
#7.	List the universities having tuition fee>15000$ and rank between 120 to 170
df['tfee']=m
df[(df['tfee']>15000) & (df['Rank']>120) & (df['Rank']<170)]
```

```python
#8.	Find the campuses of universities located in different cities (multiple cities)
df[['name', 'unicity']] = df['Name'].str.split('--', 1, expand=True)
print(df[['name', 'unicity']])
df[df["name"].duplicated()==True]["name"]
```

```python
#9.	Mention the states where out-of state fee is more than in-state students. Print Minimum and Maximum fees.
df['iee']=n
city=[]
g=list(df['Location'])
for i in g:
  k=i.split(",")[0]
  city.append(k)
df['city']=city
k=df['city'][m>n]
print(k)
p=df['tfee'].max()
q=df['tfee'].min()
print(p)
print(q)
```

```python
#10.	Find the cities locating top 100 universities.
df['city'][(df['Rank']>1) & (df['Rank']<100)]
```

```python
#11.	Find universities with least no of undergraduate students &highest no of graduate students
print(df['Undergrad Enrollment'].str.replace(",","").astype('int64').max())
print(df['Undergrad Enrollment'].str.replace(",","").astype('int64').min())
```

```python
#12.	Identifying correlations between enrollment numbers and university rank
p=df['Undergrad Enrollment'].str.replace(",","").str.split("$",expand=True).fillna(0)
df['Ug enroll']=p.astype('int64')
k=df['Rank'].corr(df['Ug enroll'])
k
```
