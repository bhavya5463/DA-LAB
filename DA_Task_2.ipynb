```python
aggregate_view=pd.DataFrame(df.groupby(by='educational_level')[['recency','annual_income','store_purchases','online_purchases']].mean()).reset_index()
aggregate_view
```

```python
import numpy as np
pivot_table = pd.DataFrame(pd.pivot_table(df, values=['sum_purchases', 'recency'], index=['marital_status'],
                        columns=['educational_level'], aggfunc={'recency': np.mean, 'sum_purchases': np.sum}, fill_value=0)).reset_index()
pivot_table
```
