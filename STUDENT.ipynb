```python
import pandas as pd
df1=pd.read_csv('/content/CSE.csv')
df2=pd.read_csv('/content/IT.csv')
df3=pd.read_csv('/content/STUDENT.CSV')
```

```python
#1.	Combine the CSE & IT data and display the data
df2.rename(columns={"PE":"Professional Elective"},inplace=True)
df=df1.append(df2,ignore_index=True)
df
```

```python
#2.	Display all CSE students’ marks along with personal information
df4=df1.merge(df3)
df4
```

```python
#3.	Print all students Regd.No, Name and professional elective.
df5=df.merge(df3)
df5[['Regd.No','Name','Professional Elective']]
```

```python
#4.	Identify the students whose DAA marks are >18.
df5[df5['DAA']>18][['Regd.No','Name']]
```

```python
#5.	Display the names and EAMCET ranks of the students who got minimum 12 marks in all courses
df5[(df5['AFL']>=12)&(df5['CN']>=12)&(df5['DAA']>=12)&(df5['OE']>=12)&(df5['Professional Elective']>=12)][['Name','EAMCET RANK']]
```

```python
#6.	Calculate mean value of all the subject’s marks.
df5[['AFL','CN','DAA','OE','Professional Elective']].mean()
```

```python
#7.	Display the names common in both CSE & IT along with Regd.No
df5[df5['Name'].duplicated()][['Name','Regd.No']]
```

```python
#8.	Fill the missing values of the data with  average marks of the subject of specific group
df['AFL']=df['AFL'].fillna(df['AFL'].mean())
df['CN']=df['CN'].fillna(df['CN'].mean())
df['DAA']=df['DAA'].fillna(df['DAA'].mean())
df['OE']=df['OE'].fillna(df['OE'].mean())
df['Professional Elective']=df['Professional Elective'].fillna(df['Professional Elective'].mean())
df
```

```python
#9.	Divide the students into 5 groups based on average marks.
df['avg']=(df['AFL']+df['CN']+df['DAA']+df['OE']+df['Professional Elective'])/5
df['group']=pd.qcut(df['avg'],5,labels=False)
df[['Regd.No','group']]
```

```python
df3['group']=pd.qcut(df3['EAMCET RANK'],6,labels=False)
df3[['Regd.No','EAMCET RANK','group']]
```
